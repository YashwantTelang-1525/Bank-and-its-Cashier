```python
from operator import index
data1 = {'Name':['Yashwant','Soham','Rajdeep','Abhi','Pranav'],
         'Age':[21,22,21,20,23],
         'Address':['Akola','Buldana','Akola','Amravati','Shimla'],
         'Qualification':['BE','HSC','ME','MSC',"Bed"]
         }
        
data2 = {'Name': ['Jaya', 'Harish', 'Anuja', 'Tanuja', 'Sakshi'], 
         'Age':[17, 14, 12, 42, 22],
         'Address':['Nagpur', 'Kanpur', 'Allahabad','Kannada','Akola'],
         'Qualification':['Btech', 'BA', 'BSc', 'BArch', 'Btech']}

df = pd.DataFrame(data1, index=[0,1,2,3,4])
df1 = pd.DataFrame(data2, index=[5,6,7,8,9])
print(df, '\n\n', df1)
```

```python
frames=[df,df1]
result=pd.concat(frames)
print(result)
```

```python
df_new = pd.DataFrame(data1, index=[0,1,2,3,4])
df1 = pd.DataFrame(data2, index=[3,4,5,6,7])

result2 = pd.concat([df_new, df1], axis=1, join='inner') #by index
print(result2)
result2 = pd.concat([df_new, df1], axis=1, join='outer') #by index
print(result2)
```
